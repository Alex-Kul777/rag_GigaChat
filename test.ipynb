```python
from data_loader import CorpusLoader
from pathlib import Path

if True:
    
    import time
    
    print("=" * 60)
    print("Тестирование загрузчика с кэшированием")
    print("=" * 60)
    
    loader = CorpusLoader()
    pdf_dir = Path("d:/study/2026-03_RAG/data/domain_1_AI/books")
    
    if pdf_dir.exists():
        # Первый запуск - загрузка из PDF
        print("\n📁 Первый запуск - загрузка из PDF (создание кэша)...")
        start_time = time.time()
        
        documents = loader.load_from_pdf_directory(
            pdf_dir,
            recursive=True,
            chunk_size=500,
            chunk_overlap=50,
            force_reload=True  # Принудительная загрузка
        )
        
        elapsed = time.time() - start_time
        print(f"✅ Загружено {len(documents)} документов/чанков за {elapsed:.2f} сек")
        
        # Второй запуск - из кэша
        print("\n📦 Второй запуск - загрузка из кэша...")
        start_time = time.time()
        
        documents_cached = loader.load_from_pdf_directory(
            pdf_dir,
            recursive=True,
            chunk_size=500,
            chunk_overlap=50,
            force_reload=False  # Используем кэш
        )
        
        elapsed = time.time() - start_time
        print(f"✅ Загружено {len(documents_cached)} документов/чанков за {elapsed:.2f} сек")
        
        # Информация о кэше
        cache_info = loader.get_cached_info(pdf_dir)
        print(f"\n📊 Информация о кэше:")
        print(f"  Кэш создан: {cache_info.get('timestamp', 'неизвестно')}")
        print(f"  Документов в кэше: {cache_info.get('num_documents', 0)}")
        
        # Статистика
        if documents:
            stats = loader.compute_stats(documents)
            print(stats)
        
        # Пример документа
        if documents:
            sample_id = list(documents.keys())[0]
            print(f"\n📄 Пример документа ({sample_id}):")
            print(documents[sample_id][:300] + "...")
        
        # Опция очистки кэша
        print("\n💡 Для очистки кэша вызовите: loader.clear_cache(pdf_dir)")
        
    else:
        print(f"❌ Директория не найдена: {pdf_dir}")
    
    print("\n✅ Тестирование завершено")
```

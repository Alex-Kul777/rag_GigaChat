```python
from rag_core import RAGPipeline
from pathlib import Path
import json

pipeline = RAGPipeline(
    chunk_size=800,
    chunk_overlap=100,
    embedding_type="gigachat",
    llm_type="gigachat"
)

pdf_dir = Path("data/domain_2_Debug/books")
if pdf_dir.exists():
    pipeline.load_from_pdf_directory_with_metadata(pdf_dir, recursive=True, force_reload=False)

print("\n📋 Реальные ID документов:")
for i, doc_id in enumerate(list(pipeline.documents_metadata.keys())[:10], 1):
    print(f"{i}. {doc_id}")
    
# Сохраняем ID в файл
with open("real_doc_ids.json", "w", encoding="utf-8") as f:
    json.dump(list(pipeline.documents_metadata.keys()), f, indent=2, ensure_ascii=False)
print("\n✅ ID сохранены в real_doc_ids.json")
```
